# Multiple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('50_Startups.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

## Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])],remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
# no matter higher or lower values coef will compensate to put everything on same scale
# thus there is no need of feature scaling in multiple regression model.
```

Yes, the p-values associated with the coefficients in a regression model can impact the interpretation of the model and influence decisions about variable inclusion. Here are some considerations:

Variable Importance: If the p-value for a particular coefficient is very small (typically less than a chosen significance level like 0.05), it suggests that the corresponding independent variable is likely to be important in predicting the dependent variable. On the other hand, a large p-value may indicate that the variable is not statistically significant.

Model Simplification: High p-values for certain coefficients may lead you to consider simplifying your model by removing those variables. This is particularly important in situations where you want a parsimonious model that includes only the most relevant variables.

Multicollinearity: In the presence of multicollinearity (high correlation between independent variables), p-values can become unstable, and the interpretation of individual coefficients may be challenging. It's crucial to be aware of multicollinearity issues in regression analysis.

Overfitting: Including variables with high p-values might lead to overfitting, where the model fits the training data too closely, potentially reducing its generalization performance on new, unseen data.

However, it's important to note that p-values are not the only criterion for assessing the quality of a model. While they can provide insights, they should be considered alongside other factors such as effect size, domain knowledge, and the overall model performance. In some cases, researchers and practitioners are moving toward focusing on effect sizes and confidence intervals rather than solely relying on p-values for model interpretation.

```python
#selection of best p values is also taken care and  backward tracking of the dataset is also taking care in the the class.
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred=regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test), 1)),1))

```

## Making a single prediction (for example the profit of a startup with R&D Spend = 160000, Administration Spend = 130000, Marketing Spend = 300000 and State = 'California')



Therefore, our model predicts that the profit of a Californian startup which spent 160000 in R&D, 130000 in Administration and 300000 in Marketing is $ 181566,92.

**Important note 1:** Notice that the values of the features were all input in a double pair of square brackets. That's because the "predict" method always expects a 2D array as the format of its inputs. And putting our values into a double pair of square brackets makes the input exactly a 2D array. Simply put:

$1, 0, 0, 160000, 130000, 300000 \rightarrow \textrm{scalars}$

$[1, 0, 0, 160000, 130000, 300000] \rightarrow \textrm{1D array}$

$[[1, 0, 0, 160000, 130000, 300000]] \rightarrow \textrm{2D array}$

**Important note 2:** Notice also that the "California" state was not input as a string in the last column but as "1, 0, 0" in the first three columns. That's because of course the predict method expects the one-hot-encoded values of the state, and as we see in the second row of the matrix of features X, "California" was encoded as "1, 0, 0". And be careful to include these values in the first three columns, not the last three ones, because the dummy variables are always created in the first columns.

```python
print(regressor.predict([[1, 0, 0, 160000, 130000, 300000]]))
```

```python
# Scatter plot for predicted vs actual values
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual Values vs Predicted Values')
plt.show()

# Residual plot
residuals = y_test - y_pred
plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='r', linestyle='--', linewidth=2)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```
